```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
import pandas as pd

data = {
    'text': [
        "Free money offer",
        "Win cash prize now",
        "Limited time offer",
        "Hello how are you",
        "Let's meet tomorrow",
        "Are you coming to class",
        "Free entry in contest",
        "Call me later"
    ],
    'label': [
        1, 1, 1,   # Spam
        0, 0, 0,   # Not Spam
        1, 0
    ]
}

df = pd.DataFrame(data)
print(df)

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df['text'])

y = df['label']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
model = MultinomialNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Predictions:", y_pred)
print("Actual:", y_test.values)

print("Accuracy:", accuracy_score(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
import joblib


joblib.dump(model, 'naive_bayes_model.pkl')


joblib.dump(vectorizer, 'vectorizer.pkl')

loaded_model = joblib.load('naive_bayes_model.pkl')


loaded_vectorizer = joblib.load('vectorizer.pkl')
text = ["Free offer now"]


text_vec = loaded_vectorizer.transform(text)


prediction = loaded_model.predict(text_vec)

print("Prediction:", prediction[0])


import numpy as np

from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB


from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

categories = ['sci.space', 'rec.sport.hockey']
data = fetch_20newsgroups(subset='all', categories=categories)

texts = data.data
labels = data.target

# Display sample
for i in range(3):
    print("Text:", texts[i][:200])
    print("Label:", labels[i])
    print("------")
import re

def preprocess(text):
    text = text.lower()  # lowercase
    text = re.sub(r'[^a-z\s]', '', text)  # remove punctuation
    return text

texts = [preprocess(t) for t in texts]
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(texts)

X_train, X_test, y_train, y_test = train_test_split(
    X, labels, test_size=0.2, random_state=42
)
model = MultinomialNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("Predictions:", y_pred)
print("Actual:", y_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
import joblib


joblib.dump(model, 'naive_bayes_model_news.pkl')


joblib.dump(vectorizer, 'vectorizer_news.pkl')
# Load model
loaded_model = joblib.load('naive_bayes_model_news.pkl')

# Load vectorizer
loaded_vectorizer = joblib.load('vectorizer_news.pkl')
test_samples = [
    "NASA launched a new satellite into space",
    "The hockey team won the championship game",
    "3D rendering and computer graphics techniques improved"
]

# Transform
test_vec = loaded_vectorizer.transform(test_samples)

# Predict
predictions = loaded_model.predict(test_vec)

# Map labels to category names
for text, pred in zip(test_samples, predictions):
    print("\nText:", text)
    print("Predicted Category:", data.target_names[pred])
```
